```python
import pandas as pd
import os
```

```python
def run(base_path):
    audio_path = []
    text = []
    for file in os.listdir(f"{base_path}/seperate"):
        try:
            with open(f'{base_path}/txt/{file[:-4]}.txt','r') as f:
                lines = f.read()
            lines = lines.strip()
            # print(lines)

            audio_path.append(f"{base_path}/seperate/{file}")
            text.append(lines)
        except:
            print(file)
            pass

    df = pd.DataFrame({'file_path':audio_path,'transcript':text})
    df.to_csv(f'{base_path}/bucket_sep.csv',index=None)
```

```python
# language = ['Hindi',"Marathi",'Bengali','Tamil','Telugu','Kannada','Malayalam']
language = ['Hindi']

base_path = '/hdd2/Kumud/SPNI_Eval_Data/Avrodh_S01_1'

for lang in language:
    run(f"{base_path}/{lang}")
```

```python
import openunmix
from openunmix import predict
import torch
import librosa
import soundfile as sf
 
def speech_source_separation(input_audio_path, output_path, model_path='umxhq'):
    # Separate sources from the input audio
    input_audio, sr = librosa.load(input_audio_path, sr=None)
    sources = predict.separate(torch.as_tensor(input_audio).float().squeeze(0), 16000)
 
    # Save the separated sources
    for source_name, source_tensor in sources.items():
        source_path = f"{source_name}.wav"
        sf.write(source_path, source_tensor.numpy(), 16000)
 
 
if __name__ == "__main__":
    # Replace 'input_audio.wav' with the path to your mixed audio file
    #input_audio_path = '/hdd/kumud/spk_id/hbj_data/SPNI_data/Sab_JholMaal_Hai_0001-B_Unmix_HD_Fun_Fair_Proxy.wav'
    input_audio_path = '/hdd2/Kumud/SPNI_Eval_Data/Avrodh_S01_1/Hindi/wav/Avrodh_0001_HD_AVC_Rev_00:02:19.57_00:02:25.35.wav'
    # Replace 'output_folder' with the desired output folder
    output_folder = '/hdd2/Kumud/SPNI_Eval_Data/Avrodh_S01_1/Hindi/seperate'
 
    # Call the separation function
    speech_source_separation(input_audio_path, output_folder)
```
