```python
from transformers import WhisperTokenizer

tokenizer = WhisperTokenizer.from_pretrained("openai/whisper-medium", task="transcribe",cache_dir='/hdd/Gothi_raj/HF_model',language = 'en')

labels = tokenizer("I love my country").input_ids
    
print(labels)
for l in labels:
    print(tokenizer.decode([l]))
print('tokenized sentence length with special tokens',len(labels))
print('tokenized sentence length',len(labels)-5)
```

```python
from transformers import WhisperTokenizer

tokenizer = WhisperTokenizer.from_pretrained("openai/whisper-medium", task="transcribe",cache_dir='/hdd/Gothi_raj/HF_model',language = 'en')
our_tokenizer = WhisperTokenizer.from_pretrained("trained_model/Alltokenized_Medium_250", task="transcribe",cache_dir='/hdd/Gothi_raj/HF_model',language = 'hi')

labels = tokenizer("मुझे अपने देश से प्यार है").input_ids
    
print("Pretrained: ")
print(labels)
for l in labels:
    print(tokenizer.decode([l]))
print('tokenized sentence length with special tokens',len(labels))
print('tokenized sentence length',len(labels)-5)


labels = our_tokenizer("मुझे अपने देश से प्यार है").input_ids

print("Our tokenizer: ")
print(labels)
for l in labels:
    print(our_tokenizer.decode([l]))
print('tokenized sentence length with special tokens',len(labels))
print('tokenized sentence length',len(labels)-5)
```
